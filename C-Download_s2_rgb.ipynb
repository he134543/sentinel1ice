# Based on the system id, download the sentinel-2 RGB images

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import geemap
import ee
from p_tqdm import p_map
```

```python
ee.Authenticate()
ee.Initialize(project="ee-sarice")
```

```python
# load sentinel-2 image collection
s2 = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
# load training lakes
training_lakes = ee.FeatureCollection("projects/ee-sarice/assets/traing_lakes_s1_s2")
# training lake ids
lake_ids = pd.read_csv("data/training_lakes.csv").Hylak_id.to_list()
# where the match tables are saved
match_table_dir = "data/match_tables"
# directory to save the downloaded RGB tifs
save_dir = "/home/xinchenh/LData/sentinel2_images/rgb"
# image scale
scale = 30 # meter
```

```python
for lake_id in lake_ids:
    
    # acquire the sentinel-2 system ids that matched with sentinel-1 images within 12 hours
    s2_ids = pd.read_csv(f"{match_table_dir}/{lake_id}_match_table.csv").s2_system_ids.unique().tolist()
    
    # check if this lake has completed
    try:
        finished_images = glob.glob(f"{save_dir}/{lake_id}/*.tif")
        if len(finished_images) == len(s2_ids):
            print("Finished")
            # finished
            continue
        else:
            pass
    except:
        # no folder established
        # not started yet
        pass
    
    # get the extent     
    # buffer 1km
    region = training_lakes.filter(ee.Filter.eq('Hylak_id', lake_id)).geometry().buffer(1000).bounds()

    
    
    # sentinel-2 image col
    rgb_col = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
                .filter(ee.Filter.inList('system:id',s2_ids)) \
                .select(["B4", "B3", "B2"]) \
                .map(lambda img: img.divide(10000))
    
    # create a directory for this lake
    out_dir = f"{save_dir}/{lake_id}"
    try:
        os.mkdir(out_dir)
    except:
        pass
    
    # download the image collection
    # make the scale as 30 to speed up the download
    geemap.download_ee_image_collection(rgb_col, 
                                    out_dir = out_dir,
                                    region = region,
                                    dtype = "float32",
                                    num_threads = 16,
                                    scale = scale,
                                    overwrite = True,
                                    )
```
